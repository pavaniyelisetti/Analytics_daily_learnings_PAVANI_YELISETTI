```python
#chapter 4: for and while loops
#creating a dataframe
import pandas as pd

cars = pd.DataFrame({
    "country": ["United States", "Australia", "Japan", "India", "Russia"],
    "cars_per_cap": [809, 731, 588, 18, 200]
}, index=["US", "AUS", "JPN", "IND", "RUS"])

cars
```

```python
#looping over dataframes
for lab, row in cars.iterrows():
    print(lab)
    print(row)
```

```python
#looping over dataframes in a format
for lab, row in cars.iterrows():
    print(lab + ": " + str(row["cars_per_cap"]))
```

```python
#adding columns using loop
for lab, row in cars.iterrows():
    cars.loc[lab, "COUNTRY"] = row["country"].upper()

print(cars)
```

```python
load = 100
hour = 0

while load < 130:
    load += 5
    hour += 1
    print(f"Hour {hour}: Load = {load}")
```

```python
# while loop with conditional logic
load = 100

while load < 150:
    if load < 120:
        status = "Low"
    elif load < 140:
        status = "Medium"
    else:
        status = "High"

    print(f"Load = {load}, Category = {status}")
    load += 10
```

```python
temps = [65, 68, 72, 75, 80]

for t in temps:
    print(f"Temperature: {t}°F")
for i, t in enumerate(temps):
    print(f"Index {i}, Temperature {t}°F")
```

```python
#vectorized alternative instead of loop by using apply
cars["COUNTRY"] = cars["country"].apply(str.upper)
print(cars)
```

```python
#numpy loop examples
import numpy as np

np_height = np.array([72, 75, 68, 80])

for x in np_height:
    print(str(x) + " inches")
```

```python
np_baseball = np.array([
    [180, 78],
    [215, 74],
    [210, 72]
])

for y in np.nditer(np_baseball):
    print(y)
```
