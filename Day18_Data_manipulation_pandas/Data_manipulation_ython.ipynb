```python
import pandas as pd
#creating a sample data
homelessness = pd.DataFrame({
    "state": ["California", "Texas", "New York", "Florida", "Colorado", "Oregon"],
    "region": ["Pacific", "West South Central", "Mid-Atlantic", "South Atlantic", "Mountain", "Pacific"],
    "individuals": [151278, 27000, 92000, 28000, 10500, 14500],
    "family_members": [34500, 8900, 18000, 12000, 950, 800]
})

homelessness
```

```python
#inspecting the dataframe
# First few rows
homelessness.head()

# Info about columns and missing values
homelessness.info()

# Shape (rows, columns)
homelessness.shape

# Summary statistics
homelessness.describe()
```

```python
#Dataframe attributes
# Values as NumPy array
homelessness.values

# Column names
homelessness.columns

# Row index
homelessness.index
```

```python
#Sorting data ascending and descending
homelessness_ind = homelessness.sort_values("individuals")
homelessness_ind.head()
homelessness_fam = homelessness.sort_values("family_members", ascending=False)
homelessness_fam.head()
```

```python
homelessness_reg_fam = homelessness.sort_values(
    ["region", "family_members"],
    ascending=[True, False]
)

homelessness_reg_fam
```

```python
#subsetting columns
individuals = homelessness["individuals"]
individuals

state_fam = homelessness[["state", "family_members"]]
state_fam

ind_state = homelessness[["individuals", "state"]]
ind_state
```

```python
#subsetting or filtering rows
ind_gt_10k = homelessness[homelessness["individuals"] > 10000]
ind_gt_10k

mountain_reg = homelessness[homelessness["region"] == "Mountain"]
mountain_reg

fam_lt_1k_pac = homelessness[
    (homelessness["family_members"] < 1000) &
    (homelessness["region"] == "Pacific")
]

fam_lt_1k_pac
```
