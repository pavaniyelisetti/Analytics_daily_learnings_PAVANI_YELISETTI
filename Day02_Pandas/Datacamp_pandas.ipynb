```python
#chapter 2; Dictionaries and pandas
#creating dictionary
import pandas as pd

#daily load data(non_confidential)
load_dict = {
    "day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "avg_temp_F" : [65, 68, 70, 72, 75, 78, 80],
    "peak_load_MW": [120, 125, 130, 128, 135, 140, 138],
    "day_type": ["Weekday", "Weekday", "Weekday", "Weekday", "Weekday", "Weekend", "Weekend"]
}
```

```python
#Accessing different dictionary elements
print(load_dict["avg_temp_F"])
print(load_dict["peak_load_MW"][3]) #3 is index for thursday
```

```python
#Dictionary manipulation
#adding a new key
load_dict["region"] = ["North"]*7

#updating an existing value
load_dict["peak_load_MW"][4] = 137 #updating friday load from 140 to 137

#removing a key
load_dict.pop("region")
```

```python
#creating nested dictionary
load_dict_nested = {
    "Mon": {"temp":65, "load":120},
    "Tue": {"temp":68, "load":125},
    "Wed": {"temp":70, "load":130}
}

print(load_dict_nested["Tue"]["load"])
```

```python
#converting dictionary to pandas dataframe
df = pd.DataFrame(load_dict)
print(df)
```

```python
#dataframe to csv & reading csv file using pandas
df.to_csv("temporary_daily_load.csv", index=False)

df_csv = pd.read_csv("temporary_daily_load.csv")
print(df_csv)
```

```python
#selection or filtering columns
#single column selection
print(df["peak_load_MW"])

#multiple columns
print(df[["day", "avg_temp_F"]])

#conditional filtering
print(df[df["avg_temp_F"] > 72])
```

```python
#loc : Label based selection
#row selection by condition
print(df.loc[df["day_type"] == "Weekend"])

#specify rows and columns
print(df.loc[0:3, ["day", "peak_load_MW"]]) #selects 0 to 3 rows of given columns

print(df.loc[:, ["day"]]) #selects all the rows of a specified column
```

```python
#iloc : index based selection
#first four rows of all columns
print(df.iloc[0:4])
#specific row and column position
print(df.iloc[1, 1])
```
